```python
import random
import pandas as pd
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data = pd.get_dummies(data, columns=['whoAmI'], drop_first= True) # с помощью get_dummies
data.head(20)
```

```python
# с помощью цикла перед заливанием в базу данных
import random
import pandas as pd
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
for i in range(len(lst)):
  if lst[i] == 'robot': lst[i] = 0
  else: lst[i] = 1
data = pd.DataFrame({'whoAmI':lst})
data.head(20)
```

```python
# уже после заливания в базу данных
import random
import pandas as pd
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst}) # исходные данные
data.loc[(data['whoAmI'] == 'robot'), 'whoAmI'] = 0
data.loc[(data['whoAmI'] == 'human'), 'whoAmI'] = 1
data.head(20)
```
